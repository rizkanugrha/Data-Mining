```python
import pandas as pd

# Load the uploaded Excel file
file_path = '/content/dataKasus-1.xlsx'

# Read the data from the Excel file
data = pd.read_excel(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Drop unnecessary columns (like Unnamed columns) and rows with missing values
data_cleaned = data.drop(columns=['Unnamed: 12'])

# Standardize column names to make them easier to work with
data_cleaned.columns = data_cleaned.columns.str.strip().str.replace(' ', '_').str.upper()

# Checking for missing values and handling them
data_cleaned = data_cleaned.dropna()

# Display the cleaned data
data_cleaned.head()
```

```python
# Creating subsets of data for Hypertension and PE/Non PE status

# Subset 1: Individuals with or without hypertension
hypertension_data = data_cleaned[['NAMA', 'RIW_HIPERTENSI']].copy()
hypertension_data['HIPERTENSI_STATUS'] = hypertension_data['RIW_HIPERTENSI'].apply(lambda x: 'Hypertensive' if x == 'Ya' else 'Non-Hypertensive')

# Subset 2: Individuals with PE or Non PE status
pe_data = data_cleaned[['NAMA', 'PE/NON_PE']].copy()
pe_data['PE_STATUS'] = pe_data['PE/NON_PE'].apply(lambda x: 'PE' if x == 'PE' else 'Non PE')

# Display the subsets
hypertension_data.head(), pe_data.head()
```
